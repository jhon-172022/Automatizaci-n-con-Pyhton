## Posiciones de Argumentos

```python
#1era Manera
def saludar(nombre, mensaje):
    print(f"{mensaje}, {nombre}")
```

```python
saludar("Carmen", "Holiwi")
```

```python
#2da Manera
def saludo(nombre, mensaje):
    print(f"Tu nombre es {nombre} y el mensaje que transmites es: {mensaje}")
```

```python
saludo(mensaje="Como te fue hoy", nombre="Carmen Millones")
```

```python
#3era Manera
def saludamos(nombre, mensaje, fecha):
    print(f"Hola {nombre}, hoy es: {fecha} y tu mensaje es: {mensaje}")
```

```python
saludamos("Carmen Millones", fecha="22/01/2024", mensaje="Practicando Python")
```

## Funciones Recursivas

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
print(factorial(2))
```
