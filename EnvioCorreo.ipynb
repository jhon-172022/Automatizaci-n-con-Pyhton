```python
import smtplib
import ssl
from email.message import EmailMessage
import imghdr
```

```python
# Define email emisor y receptor
email_emisor = 'jhonjimenez1521@gmail.com'
email_contrasena = 'zosg kise umjq xhuf'
email_receptor = 'jhoncr.jimenez@telefonica.com'
```

```python
# Fijar el asunto y el cuerpo del correo
asunto = 'Revisa'
cuerpo = """
He publicado un nuevo video en YouTu
"""

em = EmailMessage()
em['From'] = email_emisor
em['To'] = email_receptor
em['Subject'] = asunto
em.set_content(cuerpo)
```

```python
# Agregar SSL (capa de seguridad)
contexto = ssl.create_default_context()
```

```python
# Loguearse y enviar el correo
with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=contexto) as smtp:
    smtp.login(email_emisor, email_contrasena)
    smtp.sendmail(email_emisor, email_receptor, em.as_string())
```

```python
# Adjuntar archivo
with open('foto.jpg', 'rb') as file:
    file_data = file.read()
    file_tipo = imghdr.what(file.name)
    file_nombre = file.name
em.add_attachment(file_data, filename=file_nombre, subtype=file_tipo, maintype='image')
```
