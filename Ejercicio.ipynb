```python
from pathlib import Path
import re

# TODO: Renombrar "gastos.csv" basado en los nombres de carpetas padre (2022, meses, dias). El formato final seria "gastosYYYYMMDD.csv"
carpeta = Path('2022')
paths = carpeta.glob('**/*.csv')
for path in paths:
    print(path.parts)
    dia = path.parts[-2]
    dia_numero = int(re.findall('\d+', dia)[0])
    mes = path.parts[-3]
    mes_numero = int(re.findall('(\d+).', mes)[0])
    ano = path.parts[0]

    nuevo_nombre_archivo = path.with_name(f'gastos{ano}{mes_numero:02d}{dia_numero:02d}.csv')
    path.rename(nuevo_nombre_archivo)
```
