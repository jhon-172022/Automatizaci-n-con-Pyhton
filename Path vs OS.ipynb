### Listas

```python
# Propiedades de las listas:
# Son ordenadas, mantienen el orden en el que han sido definidas.
# Pueden ser formadas por tipos arbitrarios.
# Pueden ser indexadas con [i].
# Se pueden anidar, es decir, meter una dentro de la otra.
# Son mutables, ya que sus elementos pueden ser modificados.
# Son dinámicas, ya que se pueden añadir o eliminar elementos.
countries = ['United States', 'India', 'China', 'Brazil']
countries
```

```python
# Se puede acceder al primer elemento usando el índice [0].
countries[0]
```

```python
# Se puede acceder al último elemento usando el índice negativo [-1].
countries[-1]
```

```python
# Se puede modificar un elemento de la lista, basta con asignar con el operador = el nuevo valor.
countries[2] = 'Japon'
print(countries)
```

```python
# Se puede eliminar un elemento con la funcion del
del countries[2]
print(countries)
```

```python
# Podemos crear sublistas más pequeñas de una más grande [inicio:final].
Sublista = [1, 2, 3, 4, 5, 6]
print(Sublista[0:2])
print(Sublista[2:6])
```

```python
# Podemos modificar multiples valores de la lista a la vez.
ModLista = [1, 2, 3, 4, 5, 6]
ModLista[0:3] = [0, 0, 0]
print(ModLista)
```

```python
# Podemos usar el operador (+) sobre la lista
SumLista = [1, 2, 3]
SumLista += [4, 5]
print(SumLista)
```

### Listas Anidadas

```python
# Podemos tener una lista dentro de otra lista y a su vez dentro de otra lista.
# Para acceder a sus elementos sólo tenemos que usar [] tantas veces como niveles de anidado tengamos.
Lista = [1, 2, 3, ['p', 'q', [5, 6, 7]]]
print(Lista[3][0])
print(Lista[3][2][0])
print(Lista[3][2][2])
```

### Iterar Listas

```python
# En Python es muy fácil iterar una lista, mucho más que en otros lenguajes de programación.
lista = [5, 9, 10]
for l in lista:
    print(l)
```

```python
# Se pueden iterar las listas usando los índices y haciendo uso de len()
lista = [5, 9, 10]
for i in range(0, len(lista)):
    print(lista[i])
```

```python
# Si necesitamos un índice acompañado con la lista, que tome valores desde 0 hasta n-1.
lista = [5, 9, 10]
for index, l in enumerate(lista):
    print(index, l)
```

```python
# Tenemos dos listas y las queremos iterar a la vez
lista1 = [5, 9, 10]
lista2 = ["Jazz", "Rock", "Djent"]
for l1, l2 in zip(lista1, lista2):
    print(l1, l2)
```

### Métodos Listas

```python
# El método append() añade un elemento al final de la lista.
l = [1, 2]
l.append(3)
print(l)
```

```python
# El método extend() permite añadir una lista a la lista inicial.
l = [1, 2]
l.extend([3, 4])
print(l)
```

```python
# El método insert() añade un elemento en una posición o índice determinado.
l = [1, 3]
l.insert(1, 2)
print(l)
```

```python
# El método remove() recibe como argumento un objeto y lo borra de la lista.
l = [1, 2, 3]
l.remove(2)
print(l)
```

```python
# El método pop() elimina por defecto el último elemento de la lista, pero si se pasa como parámetro un índice permite borrar elementos diferentes al último.
l = [1, 2, 3]
l.pop()
print(l)
```

```python
# El método reverse() inverte el órden de la lista.
l = [1, 2, 3]
l.reverse()
print(l)
```

```python
# El método sort() ordena los elementos de menos a mayor por defecto.
l = [3, 1, 2]
l.sort()
print(l)
```

```python
# El método sort() permite ordenar de mayor a menor si se pasa como parámetro (reverse=True).
l = [3, 1, 2]
l.sort(reverse=True)
print(l)
```

```python
# El método index() recibe como parámetro un objeto y devuelve el índice de su primera aparición.
lista = ["Periphery", "Intervals", "Monuments"]
print(lista.index("Intervals"))
```

## Cambiar Nombre a Carpetas y Archivos & Obtener Ruta de todos las Subcarpetas

```python
from pathlib import Path
```

```python
Path('test').mkdir(exist_ok=True)
```

```python
# 1. Renombrar nombre carpeta
path = Path('test')
path.rename('nuevo_test')
```

```python
# 2. Renombrar archivo dentro de una carpeta
path = Path('nuevo_test/gastos-enero.txt')
nuevo_nombre_path = path.with_name('gastos-febrero.txt')
print(nuevo_nombre_path)
path.rename(nuevo_nombre_path)
```

```python
# 3. Obtener path de los subdirectorios immediatos
carpeta = Path('2022')
for path in list(carpeta.iterdir()):
    print(path)
```

```python
# 4. Obtener path de todos subdirectorios
paths = carpeta.glob('**/*')
for path in paths:
    print(path)
```

```python
# 5. Obtener path de todos los subdirectorios con un archivo
for path in paths:
    if path.is_file():
        print(path)
```

```python
# 6. Obtener path de todos los subdirectorios que tienen archivos .txt
carpeta = Path('.')
for path in carpeta.glob('**/*.txt'):
    print(path)
```

## Crear y Borrar Archivos

```python
from pathlib import Path
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
# crear archivo1, ..., archivo9 (.txt)
for i in numeros:
    with open(f'test{i}.txt', 'w') as file:
        file.write('Hola Mundo!')
```

```python
# eliminar todos archivos .txt
for path in Path('.').glob('*.txt'):
    print(path)
    path.unlink()
```

```python
# eliminar todos archivos excepto test9.txt
for path in Path('.').glob('test[1-8].txt'):
    print(path)
    path.unlink()
```
